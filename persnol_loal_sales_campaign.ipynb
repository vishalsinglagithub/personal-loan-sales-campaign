```python
X_df=data.loc[:,data.columns!="Personal Loan"] # 'columns' is the correct spelling. Also make sure "Personal Loan" matches the column name in your dataframe.
Y_df=data.loc[:,data.columns=="Personal Loan"]
```

```python
# Create the training and test data set in the ratio of 70:30 respectively. Can be any other ratio...
# Prepare data for logistic regression
from sklearn import model_selection # import the model_selection module from scikit-learn
from sklearn import metrics # import the metrics module from scikit-learn
from sklearn.metrics import confusion_matrix, classification_report
#Import LogisticRegression
from sklearn.linear_model import LogisticRegression


features=X_df.iloc[:, 0:10]
features_array = features.values   #logistic modeling algorithm requires feature array not dataframe
target_labels = Y_df.values # Changed y_data to Y_df


test_size = 0.40 # taking 60:40 training and test set
seed = 7  # Random numbmer seeding for reapeatability of the code when using random functions to
# generate the training and test data

X_train, X_test, y_train, y_test = model_selection.train_test_split(features_array, target_labels, test_size=test_size, random_state=seed)
y_train = np.ravel(y_train)   # to convert 1 d vector into 1 d array
# Let us first try logistic regression to predict the personal loan affinity -
# We removed the binned columns for this (considered only 11 columns 1:10) which are numeric

model = LogisticRegression()
model.fit(X_train, y_train)
model_score = model.score(X_test, y_test)
y_predict = model.predict(X_test) # Changed y_pred to y_predict
print(model_score)
print(metrics.confusion_matrix(y_test, y_predict)) # Now this should work


# Assuming 'y_test' are the true labels and 'y_predict' are the predicted labels # Changed y_pred to y_predict
conf_matrix = confusion_matrix(y_test, y_predict) # Changed y_pred to y_predict
class_report = classification_report(y_test, y_predict) # Changed y_pred to y_predict

print(f"Confusion Matrix:\n {conf_matrix}")
print(f"Classification Report:\n {class_report}")
```

```python
# we apply knn with minkoski n=3

from sklearn.neighbors import KNeighborsClassifier

# Create KNN classifier with Minkowski distance and n=3
knn = KNeighborsClassifier(n_neighbors=3, metric='minkowski')

# Fit the model to the training data
knn.fit(X_train, y_train)

# Predict on the test data
y_predict = knn.predict(X_test)

# Evaluate the model
model_score = knn.score(X_test, y_test)
print(model_score)
print(metrics.confusion_matrix(y_test, y_predict))

# Print confusion matrix and classification report
conf_matrix = confusion_matrix(y_test, y_predict)
class_report = classification_report(y_test, y_predict)

print(f"Confusion Matrix:\n {conf_matrix}")
print(f"Classification Report:\n {class_report}")
```

```python
#  apply naive bayes algorithm

from sklearn.naive_bayes import GaussianNB

# Create a Gaussian Naive Bayes model
nb_model = GaussianNB()

# Train the model
nb_model.fit(X_train, y_train)

# Make predictions
nb_predictions = nb_model.predict(X_test)

# Evaluate the model
nb_score = nb_model.score(X_test, y_test)
print(f"Naive Bayes Model Score: {nb_score}")

# Confusion matrix and classification report
nb_conf_matrix = confusion_matrix(y_test, nb_predictions)
nb_class_report = classification_report(y_test, nb_predictions)

print(f"Naive Bayes Confusion Matrix:\n {nb_conf_matrix}")
print(f"Naive Bayes Classification Report:\n {nb_class_report}")
```

```python
#  apply preprocessing model also provide the confusion matrix ,accuracy, classification report.

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score



# Apply StandardScaler to the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train the model with scaled data
model = LogisticRegression()
model.fit(X_train_scaled, y_train)

# Make predictions with scaled data
y_predict = model.predict(X_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, y_predict)
conf_matrix = confusion_matrix(y_test, y_predict)
class_report = classification_report(y_test, y_predict)

print(f"Accuracy: {accuracy}")
print(f"Confusion Matrix:\n {conf_matrix}")
print(f"Classification Report:\n {class_report}")
```

```python
#  : apply decision tree model

from sklearn.tree import DecisionTreeClassifier

# Create a Decision Tree model
dt_model = DecisionTreeClassifier(random_state=seed)

# Train the model
dt_model.fit(X_train, y_train)

# Make predictions
dt_predictions = dt_model.predict(X_test)

# Evaluate the model
dt_score = dt_model.score(X_test, y_test)
print(f"Decision Tree Model Score: {dt_score}")

# Confusion matrix and classification report
dt_conf_matrix = confusion_matrix(y_test, dt_predictions)
dt_class_report = classification_report(y_test, dt_predictions)

print(f"Decision Tree Confusion Matrix:\n {dt_conf_matrix}")
print(f"Decision Tree Classification Report:\n {dt_class_report}")
```

```python
#  : use decision tree for user interaction for we have to provide loan or not

import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Assuming you have your data loaded into a DataFrame called 'data'
# and your features and target variable are defined as 'X_df' and 'Y_df'

# Prepare data for model training
features = X_df[['Age', 'Income', 'Family', 'Education', 'Mortgage']] # Select the desired features
features_array = features.values
target_labels = Y_df.values
y_train = np.ravel(target_labels)  # Ensure target is 1D array

X_train, X_test, y_train, y_test = train_test_split(
    features_array, y_train, test_size=0.30, random_state=7
)

# Train the Decision Tree model
dt_model = DecisionTreeClassifier(random_state=7)
dt_model.fit(X_train, y_train)

# Function for user interaction and prediction
def get_user_input():
    """Gets user input for features and returns a feature vector."""
    # Replace with actual feature  s and input gathering
    age = float(input("Enter your age: "))
    income = float(input("Enter your income: "))
    family = float(input("Enter your family size: "))
    education = float(input("Enter your education level (1: Undergrad; 2: Graduate; 3: Advanced/Professional): "))
    mortgage = float(input("Enter your mortgage: "))
    # ... get other feature inputs
    user_features = [age, income, family, education, mortgage]  # Create a list of user inputs
    return user_features


def predict_loan(user_features, model):
    """Predicts whether a loan should be granted based on user input."""
    prediction = model.predict([user_features])
    if prediction[0] == 1:
        print("Loan granted.")
    else:
        print("Loan not granted.")


# Get user input5

user_features = get_user_input()

# Make a prediction using the trained model
predict_loan(user_features, dt_model)
```

```python
#  now we  apply z test p test and anova

import scipy.stats as stats

# Z-test (example: comparing mean income of loan takers vs non-takers)
loan_takers = data[data['Personal Loan'] == 1]['Income']
non_takers = data[data['Personal Loan'] == 0]['Income']

z_statistic, p_value = stats.ttest_ind(loan_takers, non_takers)
print(f"Z-test (Income): Z-statistic = {z_statistic}, p-value = {p_value}")

# T-test (example: comparing mean age of loan takers vs non-takers)
loan_takers_age = data[data['Personal Loan'] == 1]['Age']
non_takers_age = data[data['Personal Loan'] == 0]['Age']

t_statistic, p_value = stats.ttest_ind(loan_takers_age, non_takers_age)
print(f"T-test (Age): T-statistic = {t_statistic}, p-value = {p_value}")

# ANOVA (example: comparing mean income across different education levels)
# Assuming 'Education' is a categorical variable representing education levels
# You can replace 'Education' with your actual categorical variable
income_by_education = data.groupby('Education')['Income'].apply(list)
f_statistic, p_value = stats.f_oneway(*income_by_education)
print(f"ANOVA (Income by Education): F-statistic = {f_statistic}, p-value = {p_value}")
```

#                               ADDITIONAL ANALYSIS

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

data = pd.read_csv("Bank_Personal_Loan_Modelling.csv")

# ANOVA (example: comparing mean income across different education levels)
income_by_education = data.groupby('Education')['Income'].apply(list)
f_statistic, p_value = stats.f_oneway(*income_by_education)
print(f"ANOVA (Income by Education): F-statistic = {f_statistic}, p-value = {p_value}")

# Example: comparing mean experience across different family sizes
experience_by_family = data.groupby('Family')['Experience'].apply(list)
f_statistic, p_value = stats.f_oneway(*experience_by_family)
print(f"ANOVA (Experience by Family): F-statistic = {f_statistic}, p-value = {p_value}")

# Example: comparing mean CCAvg across different education levels
ccavg_by_education = data.groupby('Education')['CCAvg'].apply(list)
f_statistic, p_value = stats.f_oneway(*ccavg_by_education)
print(f"ANOVA (CCAvg by Education): F-statistic = {f_statistic}, p-value = {p_value}")

# Example: comparing mean mortgage across different family sizes
mortgage_by_family = data.groupby('Family')['Mortgage'].apply(list)
f_statistic, p_value = stats.f_oneway(*mortgage_by_family)
print(f"ANOVA (Mortgage by Family): F-statistic = {f_statistic}, p-value = {p_value}")
```
